```python
import json
from collections import defaultdict
import matplotlib.pyplot as plt
import random

```

## Balance the dataset

```python
def sample_n_rows(input_list, n=1000):
    if len(input_list) <= n:
        return random.sample(input_list, len(input_list))
    else:
        return random.sample(input_list, n)
```

```python
# Load the dataset annotations
train_10class_path = '/N/project/baby_vision_curriculum/benchmarks/ssv2/annotations/train-10classes.json'
val_10class_path = '/N/project/baby_vision_curriculum/benchmarks/ssv2/annotations/validation-10classes.json'

train_out_path = '/N/project/baby_vision_curriculum/benchmarks/ssv2/annotations/train_b-10classes.json'
val_out_path = '/N/project/baby_vision_curriculum/benchmarks/ssv2/annotations/val_b-10classes.json'

c_path = val_10class_path#train_10class_path#
out_path = val_out_path

with open(c_path, 'r') as f:
    dataset = json.load(f)

print(len(dataset))

# Mapping of existing class labels to new class labels
class_numbers = defaultdict(list)

# new_data = []
# Convert class labels
for data in dataset:
    class_numbers[data['template']].append(data)
        # new_data.append(data)
```

```python
for key in class_numbers.keys():
    print(key, len(class_numbers[key]))
```

```python
n_b_samples = 100 #1000
```

```python

# print(class_numbers)
final_li = []
for key in class_numbers.keys():
    print(key, 'num samples before balancing', len(class_numbers[key]))
    final_li.extend(sample_n_rows(class_numbers[key],
                                 n=n_b_samples))

print(final_li, len(final_li))
random.shuffle(final_li)
# Save the modified dataset
with open(out_path, 'w') as f:
    json.dump(final_li, f)

# def plot_histogram(data_dict):
#     categories = list(data_dict.keys())
#     frequencies = list(data_dict.values())

#     # Plot the histogram
#     plt.bar(categories, frequencies)
    
#     # Add labels and title
#     plt.xlabel('Categories')
#     plt.ylabel('Frequency')
#     plt.title('Histogram')
#     plt.xticks(categories, rotation='vertical')

#     # Show the plot
#     plt.show()

# # Call the function with your data dictionary
# plot_histogram(class_numbers)
```

## Convert to CSV

```python
import json
import csv
```

```python
input_json_file = '/N/project/baby_vision_curriculum/benchmarks/ssv2/annotations/val_ann_b-10classes.json'
output_csv_file = '/N/project/baby_vision_curriculum/benchmarks/ssv2/easy_labels/val_easy10.csv'
```

```python
# Read JSON data from the file
with open(input_json_file, 'r') as f:
    data = json.load(f)

# Prepare the list of dictionaries for CSV writing
rows = [{"fname": entry["id"] + ".webm", "label": entry["template"]} for entry in data]

# Write CSV data to the file
with open(output_csv_file, 'w', newline='') as f:
    fieldnames = ['fname', 'label']
    writer = csv.DictWriter(f, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(rows)

print(f"Modified CSV file '{output_csv_file}' has been created successfully.")
```
